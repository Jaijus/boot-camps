Lab notebook

Increase transparency and verifiability 

Managing large data (see: [http://git-annex.branchable.com/](http://git-annex.branchable.com/))

Lowering barriers to reuse

## Getting git

Git ships with many operating systems. Try: 

```python
!git
```

If you get an error:
    
`-bash: git: command not found`

You will need to install git. 

On linux, that's easy. On ubuntu and similar distro's:  

`$ sudo apt-get install git`

On Fedora, 

`$ sudo yum install git`

If you are on other OS, it's easy to get what you need from: [http://git-scm.com/downloads](http://git-scm.com/downloads)

You can get help by typing:

```python
!git help
```

```python
!git help commit
```

The first time you start to use git, you'll need to configure your author information:

`$ git config --global user.name 'Ariel Rokem'`

`$ git config --global user.email arokem@stanford.edu`



You'll want to make sure that your editor is configured: 

`$ git config --global core.editor emacs`
